```python
import os

folder_path = r'ds003688\ds003688\stimuli\annotations\sound'  

file_list = os.listdir(folder_path)

labels_df = pd.DataFrame(dict())
for file_name in file_list:
    df = pd.read_csv(fr'ds003688\ds003688\stimuli\annotations\sound\{file_name}', delimiter='\t')
    for i in range(len(df)):
        y = np.zeros(4216, dtype=int)
        y[np.arange(int(df['onset'][i]*10),ceil(df['offset'][i]*10)+1)] = 1
        labels_df[f"{df['item'][i]}_{file_name[17:-3]}"] = y
labels_df
```
